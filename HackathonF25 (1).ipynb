```python
%%writefile ecoBuy.py
import streamlit as st
import google.generativeai as genai

GOOGLE_API_KEY = ""

genai.configure(api_key=GOOGLE_API_KEY)
#model = genai.GenerativeModel("gemini-2.5-pro")

PREPROMPT = {"role": "user", "parts": ["You are an educational grocery assistant called EcoBuy. For each concern given by the user, your job is to first say Yes/No depending on whether the concern is valid for the specific product mentioned (the format for this response should be CONCERN NAME: YES/NO). On the next line, if and only if the answer to the previous part has a NO, list a few specific alternatives for the product under the title- Alternative Options. Give only bullet points"] }

st.image("https://thumbs.dreamstime.com/b/supermarket-interior-flat-vector-illustration-grocery-store-shelves-food-products-cartoon-food-shop-aisle-bakery-meat-166466892.jpg")

custom_title = """
<style>
.custom-title {
    font-family: 'Arizona', serif; /* Specify your desired font family */
    color: #FF5733; /* Set the color */
    font-size: 48px; /* Adjust the font size */
    text-align: center; /* Align the text */
}
</style>
<h1 class="custom-title">EcoBuy ðŸ›’</h1>
"""
st.markdown(custom_title, unsafe_allow_html=True)


#  st.title("EcoBuy ðŸ›’")


st.set_page_config(
   page_title="EcoBuy",
   page_icon="ðŸŒ±",
   layout="centered"
)
st.write("With EcoBuy, you can uncover the environmental impact behind the products you buy every day. Simply choose a category, like groceries, personal care, or household, and enter an item to learn how sustainable it really is. Whether youâ€™re checking if somethingâ€™s recyclable, biodegradable, or ethically made, EcoBuy helps you make informed, eco-conscious choices that create a greener future.")

st.markdown(
   """
    <style>
    [data-testid="stAppViewContainer"] {
        background-color: #8da68d;
    }
    </style>
    """,
   unsafe_allow_html=True
)

options = [" ", "Groceries", "Household", "Personal care"]
selectedOption= st.selectbox("Select a category:", options)

if selectedOption == "Groceries":
  #st.markdown(":green[Please select all the concerns that apply].")
  concerns= ["Pesticide-free",
            "Vegan",
            "Low carbon footprint"]
  concernOptions = st.multiselect("Select one or more concerns",concerns)
  product= st.text_input("Enter your product:")


elif selectedOption == "Household":
  #st.markdown(":green[Please select all the concerns that apply].")
  concerns= ["Biodegradable",
            "Cruelty-free",
            "Microplastics-free"]
  concernOptions = st.multiselect("Select one or more concerns",concerns)
  product= st.text_input("Enter your product:")


else:
  #st.markdown(":green[Please select all the concerns that apply].")
  concerns= ["Biodegradable",
            "Cruelty-free",
            "Microplastics-free"]
  concernOptions = st.multiselect("Select one or more concerns",concerns)
  product= st.text_input("Enter your product:")

user_input = f"I am buying a {product} and I want to know if the following concerns are related to this product: {', '.join(concernOptions)}"
geminiModel=genai.GenerativeModel("gemini-2.5-pro")
chat = geminiModel.start_chat(history=[PREPROMPT])


def get_gemini_response(query):
  instantResponse=chat.send_message(query,stream=False)
  return instantResponse

col1, col2, col3 = st.columns([0.3, 0.6, 0.1])

if st.button("Submit"):
  output=get_gemini_response(user_input)
  with col2:
    st.subheader(output.text)
```
